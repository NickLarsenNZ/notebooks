## Write a LaTeX mathematical expression

$(a + b)^2 = a^2 + 2ab + b^2$

```python
%%manim Equation
from manimlib.imports import *

class Equation(Scene):
    def construct(self):
        equation = TexMobject(
            "(",  # 0
            "a",  # 1
            "+",  # 2
            "b",  # 3
            ")^", # 4
            "2",  # 5
            
            "=",  # 6 
            
            "a^",  # 7
            "2", # 8
            "+",  # 9
            "2",  # 10
            "a",  # 11
            "b",  # 12
            "+",  # 13
            "b^",  # 14
            "2", # 15
        )
        
        equation.move_to(2*UP)
        
#         equation[7].set_color(BLUE)
        
        equation[1].set_color(RED_B)
        equation[7].set_color(RED_B)
        equation[11].set_color(RED_B)
        equation[3].set_color(BLUE_B)
        equation[12].set_color(BLUE_B)
        equation[14].set_color(BLUE_B)
                              
        equation.scale(1.5)
        self.play(Write(equation[0:7]))
        self.wait()
        self.play(
            ReplacementTransform(equation[1].copy(),equation[7]),
            ReplacementTransform(equation[3].copy(),equation[14]),
            ReplacementTransform(equation[5].copy(),equation[8]),
            ReplacementTransform(equation[5].copy(),equation[15]),
            ReplacementTransform(equation[2].copy(),equation[9]),
            ReplacementTransform(equation[2].copy(),equation[13]),
            run_time=1.5
        )
        self.wait()
        self.play(
            ReplacementTransform(equation[1].copy(),equation[11]),
            ReplacementTransform(equation[3].copy(),equation[12]),
            ReplacementTransform(equation[2].copy(),equation[10]),
            run_time=1.5
        )
        self.wait()
        
        a = 1
        b = 2
        
        a_plus_b_square = Square(
            side_length = a+b, 
            fill_color=WHITE, 
            fill_opacity=0.5, 
            color=WHITE
        )
        a_plus_b_square.move_to(DOWN)
        
        self.play(
            ReplacementTransform(equation[0:6].copy(),a_plus_b_square)
        )
        self.wait()
        equation[11].set_color(YELLOW)
        equation[12].set_color(YELLOW)
        
        a_square = Square(
            side_length = a, 
            fill_color=RED_B, 
            fill_opacity=0.5, 
            color=RED_B
        )
        a_square.move_to(a_plus_b_square.get_corner(DOWN+LEFT),DOWN+LEFT)
        
        b_square = Square(
            side_length = b, 
            fill_color=BLUE_B, 
            fill_opacity=0.5, 
            color=BLUE_B
        )
        b_square.move_to(a_plus_b_square.get_corner(UP+RIGHT),UP+RIGHT)
        
        ab_rectangle = Rectangle(
            width = b,
            height = a,
            fill_color=YELLOW, 
            fill_opacity=0.5,
            color=YELLOW
        )
        ab_rectangle.move_to(a_plus_b_square.get_corner(DOWN+RIGHT),DOWN+RIGHT)
        
        ba_rectangle = Rectangle(
            width = a,
            height = b,
            fill_color=YELLOW, 
            fill_opacity=0.5,
            color=YELLOW
        )
        ba_rectangle.move_to(a_plus_b_square.get_corner(UP+LEFT),UP+LEFT)
        
        self.play(
            ReplacementTransform(equation[7].copy(),a_square),
            ReplacementTransform(equation[8].copy(),a_square),
            run_time=1.5
        )
        self.play(
            ReplacementTransform(equation[14].copy(),b_square),
            ReplacementTransform(equation[15].copy(),b_square),
            run_time=1.5
        )
        self.play(
            ReplacementTransform(equation[11].copy(),ab_rectangle),
            ReplacementTransform(equation[12].copy(),ba_rectangle),
            run_time=1.5
        )
```
